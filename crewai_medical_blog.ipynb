```python
from crewai import Agent,Crew,Task
```

```python
from langchain_groq import ChatGroq
from crewai.tools import tool
from langchain_community.tools.tavily_search import TavilySearchResults
import os
from google.colab import userdata
tavily_api_key=userdata.get('tavily_api_key')
groq_api_key=userdata.get('groq_api_Key')
```

```python
os.environ["GROQ_API_KEY"] = groq_api_key
```

```python
llm=ChatGroq(
    model_name="groq/llama3-8b-8192",temperature=0.7
)
```

```python
from langchain_community.utilities.tavily_search import TavilySearchAPIWrapper # import tavily search API wrapper
```

```python
tool=TavilySearchResults(api_wrapper=TavilySearchAPIWrapper(tavily_api_key=tavily_api_key))
```

```python
# Researcher Agent - Finds latest medical data
researcher = Agent(
    role="Medical Researcher",
    goal="Find the latest scientific data on AI in healthcare",
    backstory="A medical expert specializing in digital health innovations",
    llm=llm,
    tools=[tool]
)
```

```python
# Writer Agent - Creates SEO-optimized articles
writer = Agent(
    role="Medical Content Writer",
    goal="Write detailed, engaging, and SEO-optimized healthcare articles",
    backstory="A professional medical journalist and AI blogger",
    llm=llm,
    tools=[tool]
)
```

```python
# Writer Agent - Creates SEO-optimized articles
writer = Agent(
    role="Medical Content Writer",
    goal="Write detailed, engaging, and SEO-optimized healthcare articles",
    backstory="A professional medical journalist and AI blogger",
    llm=llm
)
```

```python
# Editor Agent - Fact-checks and enhances readability
editor = Agent(
    role="Medical Editor",
    goal="Fact-checks AI-written content and improves readability",
    backstory="A medical proofreader and AI optimization expert",
    llm=llm
)
```

```python
# SEO Expert - Optimizes for search engines
seo_expert = Agent(
    role="SEO Specialist",
    goal="Optimize the blog for search engines with keywords and structure",
    backstory="An expert in content marketing and SEO in healthcare",
    llm=llm
)
```

```python
# Social Media Agent - Creates Twitter/LinkedIn posts
social_media = Agent(
    role="Social Media Manager",
    goal="Write engaging LinkedIn and Twitter posts summarizing the blog",
    backstory="An AI-driven social media expert",
    llm=llm
)
```

```python
import time
research_task = Task(description="Find recent AI in healthcare trends from PubMed and research papers.",
                     expected_output= "A comprehensive list of recent trends and research papers on AI in healthcare.",
                     agent=researcher)
writing_task = Task(description="Write a 1500-word article based on the research.",
                    expected_output="A well-written and detailed 1500-word article on recent AI trends in healthcare.",
                    agent=writer)
editing_task = Task(description="Proofread, fact-check, and improve clarity.",
                    expected_output="A proofread, fact-checked, and clear article ready for publishing.",
                    agent=editor)
editing_task = Task(description="Proofread, fact-check, and improve clarity.",
                    expected_output="A proofread, fact-checked, and clear article ready for publishing.",
                    agent=editor)
seo_task = Task(description="Add keywords, meta descriptions, and internal links.",
                expected_output="An SEO-optimized article with relevant keywords, meta descriptions, and internal links.",
                agent=seo_expert)
social_task = Task(description="Create a LinkedIn & Twitter summary of the blog.",
                  expected_output="Engaging LinkedIn and Twitter posts summarizing the key points of the blog.",
                   agent=social_media)

crew = Crew(agents=[researcher, writer],
            tasks=[research_task, writing_task],verbose=1)

result = crew.kickoff()
# print the output in markdown file
print("Waiting 20 seconds before printing markdown...")
time.sleep(20)

print(result)
```
